```python
import pandas as pd
```

```python
data_path = "data\PDFEinfo - Formatted.csv"
df = pd.read_csv(data_path)

print(df)
```

```python
df.tail()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.describe()
```

```python
filter1 = (df['sessions #'] == 1)
dfs1 = df[filter1]
print(dfs1)

```

```python
dfs1['Session 1 - time of FoG (s)'].info()

dfs1 = dfs1.select_dtypes(exclude = ['object'])
```

```python
dfs1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import BayesianRidge
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

input = dfs1.dropna()
output = dfs1['Session 1 - H&Y (score)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, output, test_size = 0.2, random_state = 42)
```

```python
logred_model = LogisticRegression()
logred_model.fit(X_train, y_train)

knn_model = KNeighborsClassifier(n_neighbors = 3)
knn_model.fit(X_train, y_train)

bayesian_model = BayesianRidge()
bayesian_model.fit(X_train, y_train)

rf_model = RandomForestClassifier(n_estimators = 100)
rf_model.fit(X_train, y_train)
```

```python
y_pred = logred_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
y_pred_bayesion = bayesian_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Logistic Regression Accuracy: {accuracy}')

accuracy = accuracy_score(y_test, y_pred_knn)
print(f'KNN Accuracy: {accuracy}')

accuracy = accuracy_score(y_test, y_pred_rf)
print(f'Random Forest Accuracy: {accuracy}')
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred_bayesion)
print(f'Bayesion MAE: {mae}')
```
