# Data Analysis 

````python
# Package for reading .csv files
import pandas as pd

# Package for computing matrices
import numpy as np
````

````python
# Define the path of the data
data_path = "data/PDFEinfo - Formatted.csv"

# Load the .csv file into a variable
df = pd.read_csv(data_path, encoding = 'latin-1')

print(df)
````

### Viewing the Dataframe

There are a few functions we can use to view parts of the dataframe
1. ``` df.head() ``` Shows the top 5 rows of your dataframe
2. ``` df.tail() ``` Shows the bottom 5 rows of your dataframe
3. ``` df.describe() ``` Provides common metrics of your entire dataframe
4. ``` df.info() ``` Provides index, column name, Non-null count and datatype information

````python
df.info()
````

Notice how some of the datatypes are of type 'object'. When looking at each individual cell, we notice that there is no data in it. Some datatypes of NAN are seen as float64 in the code. To remove unwatned values that can skew the data, we can use ``` df.dropna() ``` to remove the NAN values

````python
df = df.dropna()
````
